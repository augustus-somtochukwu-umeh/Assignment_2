```python
import pandas as pd
import numpy as np
```

```python
#load dataset
food_choices = pd.read_csv("food_coded.csv")
food_choices.head()
```

```python
food_choices.info()
```

```python
food_choices.isnull().sum().sort_values(ascending=False).head(5)
```

```python
food_choices["GPA"] = food_choices["GPA"].replace(r'^\s*$', np.nan, regex=True)
food_choices["GPA"].isnull().sum()
```

```python
food_choices["GPA"] = pd.to_numeric(food_choices["GPA"], errors="coerce")
type(food_choices["GPA"])
food_choices["GPA"].isnull().sum()
```

```python
cols = ["GPA", "weight", "fruit_day"]
food_choices[cols] = food_choices[cols].apply(pd.to_numeric, errors="coerce")
food_choices.info()
```
