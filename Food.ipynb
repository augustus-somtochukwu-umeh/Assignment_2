```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
food_choice = pd.read_csv("food_coded.csv")
food_choice.head(10)
```

```python
food_choice.info()
```

```python
food_choice.isnull().sum().sort_values(ascending = False)
```

```python
food_choice['GPA'] = food_choice['GPA'].replace("", np.nan)
food_choice['GPA'] = pd.to_numeric(food_choice['GPA'], errors='coerce')
food_choice['GPA'] = food_choice['GPA'].replace("", np.nan)
food_choice['GPA'] = food_choice['GPA'].fillna(food_choice['GPA'].median())
food_choice['GPA'].info()
```

```python
food_choice['weight'] = food_choice['weight'].replace("", np.nan)
food_choice['weight'] = pd.to_numeric(food_choice['GPA'], errors='coerce')
food_choice['weight'] = food_choice['GPA'].fillna(food_choice['weight'].median())
food_choice['weight'].info()
```

```python
food_choice['fruit_day'] = pd.to_numeric(food_choice['fruit_day'], errors='coerce')
food_choice['fruit_day'].info()
```

```python
food_choice['calories_day'] = food_choice['calories_day'].fillna(food_choice['calories_day'].median())
food_choice['calories_day'].info()
```

```python
food_choice['Gender']= food_choice['Gender'].map({1: "Female", 2: "Male"})
food_choice['Gender']
```

```python
food_choice['comfort_food_reasons'] = food_choice['comfort_food_reasons'].astype(str)
food_choice['comfort_food_reasons'].info()
```

```python
food_choice['cuisine'] = pd.to_numeric (food_choice['cuisine'], errors='coerce')
food_choice['cuisine'].info()
```
